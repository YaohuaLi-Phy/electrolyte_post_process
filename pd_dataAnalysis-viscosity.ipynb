```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
```

```python
path = './viscosity/'
filename = '22-7-20-check-correct_viscosity.csv'
df = pd.read_csv(path+filename)
df.describe()
```

```python
df
```

```python
salt1 = []
salt2 = []
solvent1 = []
solvent2 = []
for i in range(df.shape[0]):
    f1=df.iloc[i][0]
    more_labels = f1.split("-")
    if len(more_labels) == 6:
        salt1.append(more_labels[2])
        salt2.append('-')
        solvent1.append(more_labels[-1])
        solvent2.append('-')
    if len(more_labels) == 7:
        salt1.append(more_labels[2])
        salt2.append('-')
        solvent1.append(more_labels[-1])
        solvent2.append(more_labels[-2])
    if len(more_labels) == 12:
        solvent1.append(more_labels[-1])
        solvent2.append('-')
        salt1.append(more_labels[6])
        salt2.append(more_labels[8])
    if len(more_labels) == 13:
        solvent1.append(more_labels[-1])
        solvent2.append(more_labels[-2])
        salt1.append(more_labels[6])
        salt2.append(more_labels[8])
        
```

```python
df['salt1'] = salt1
df['salt2'] = salt2
df['solvent1'] = solvent1
df['solvent2'] = solvent2
```

```python
sel1 = df.loc[(df['solvent1'] == 'ec') | (df['solvent2'] == 'ec')]
sel1.sort_values(by='viscosity',ascending=True)
```

```python
sel1 = df.loc[(df['solvent1'] == 'dme') | (df['solvent2'] == 'dme')]
dli = sel1['viscosity']
tli = sel1['R_values2']
nparr = dli.to_numpy()
nptli = tli.to_numpy()
plt.figure()
hist, edges = np.histogram(nptli, bins=11)
histD, edgesD = np.histogram(nparr, bins=11)
ax = tli.plot.hist(xlabel='D_Li+', bins=11)
ax.set_xlabel=("D")
```

```python
keyword = 'fsi'
sel1 = df.loc[(df['salt1'] == keyword) | (df['salt2'] == keyword)]
sel1.to_csv('select_'+keyword+'.csv')
sel1.sort_values(by='viscosity',ascending=False)
```

```python
norm_sum = sum(hist)
hist1 = hist/norm_sum
histD1 = histD/sum(histD)
x1 = (edges[1:]+edges[:-1])/2
xD1 = (edgesD[1:]+edgesD[:-1])/2
x= [x1]
xD= [xD1]
hist = [hist1]
histD = [histD1]
```

```python
plt.figure()
plt.plot(x1, hist1, label='LiPF6')
plt.legend()
```

```python
salt_names = ['pf6', 'fsi', 'dfob', 'bob', 'bf4', 'tfsi']
for ii in range(1,6):
    sel = df.loc[(df['salt1'] == salt_names[ii]) | (df['salt2'] == salt_names[ii])]
    dli = sel['D_Li+']
    tli = sel['t+']
    nparr = dli.to_numpy()
    nptli = tli.to_numpy()
    histi, edges = np.histogram(nptli, bins=15)
    histDi, edgesD = np.histogram(nparr, bins=15)
    hist.append(histi/sum(histi))
    histD.append(histDi/sum(histDi))
    x.append((edges[1:]+edges[:-1])/2)
    xD.append((edgesD[1:]+edgesD[:-1])/2)
    
```

```python
plt.figure()
for i in range(6):
    plt.plot(xD[i],histD[i],label=salt_names[i])
plt.xlabel('Diffusion $m^2$/s')
plt.ylabel('Normalized ratio')
plt.legend()
plt.savefig('diffusion_constant.png', dpi=300)
```
