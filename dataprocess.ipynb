```python
class Solution(object):
    
    def search(self, nums, target,cumulative):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        
        ldata=len(nums)
        if ldata>3:
            a = int(ldata/2)
            if target == nums[a-1]:
                return (cumulative+a-1)
            elif target == nums[a]:
                return(cumulative+a)
            elif target < nums[a-1]:
                print('left part')
                print('cumulative='+str(cumulative))
                new_nums = nums[0:a]
                print(new_nums)
                return self.search(new_nums, target, cumulative)
            elif target > nums[a]:
                print('right part')
                cumulative += a + 1 
                print('cumulative='+str(cumulative))
                new_nums = nums[a+1:]
                print(new_nums)
                
                return self.search(new_nums, target, cumulative)
        elif ldata==2:
            print('ldata=2')
            if nums[0]==target:
                return (cumulative)
            elif nums[1]==target:
                return (cumulative+1)
            else:
                return -1
        elif ldata==3:
            if nums[0]==target:
                return (cumulative)
            elif nums[1]==target:
                return (cumulative+1)
            elif nums[2]==target:
                return(cumulative+2)
            else:
                return -1
        
```

```python
sol=Solution()
```
