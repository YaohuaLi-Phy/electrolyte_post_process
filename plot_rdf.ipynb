```python
name1 = 'p002-pf6-dmc/gofr_Li-F.dat'
data = np.genfromtxt('p002-pf6-dmc/gofr_Li-F.dat')
data2 = np.genfromtxt('03-15-database/0166-pf6-fsi-dmc/gofr-Li-PF6_F.dat')
fig, ax = plt.subplots()
ax.plot(data[:, 0], data[:, 1], label='Li-P (LiPF6)')
ax.plot(data2[:, 0], data2[:, 1], label='Li-P (LiFSI and LiPF6)')
ax.set_xlabel("r (nm)")
ax.set_ylabel("g(r)")
ax2 = ax.twinx()
ax2.plot(data[:, 0], data[:, 2], label='Li-P (LiPF6)')
ax2.plot(data2[:, 0], data2[:, 2], label='Li-P (LiFSI and LiPF6)')
ax2.set_ylabel("N(r)")
ax.legend()
fig.savefig(name1+".png", dpi=300)
#ax2.set_ylim(0, 30)
```

```python
data = np.genfromtxt('p002-pf6-dmc/li-msd.xvg', skip_header=21)
print(len(data))
x = data[:-1300, 0]
y = data[:-1300, 1]
```

```python
def test(x, a, b):
    return a * x + b
```

```python
test_x = np.linspace(1, 3000,200)
param, param_cov = curve_fit(test, x, y)
param
```

```python
test_y = param[0]*test_x + param[1]
```

```python
plt.plot(data[:, 0], data[:,1], label='MSD data')
plt.plot(test_x, test_y, label='linear fitting')
plt.legend()
plt.xlabel('Simulation Time (ps)')
plt.ylabel('MSD (nm)')
plt.savefig("MSD_example.png")
```
